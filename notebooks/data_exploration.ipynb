```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('../data/synthetic_data.csv')
df['visit_date'] = pd.to_datetime(df['visit_date'])

# Plot the distribution of visits over time
plt.figure(figsize=(12, 6))
df['visit_date'].value_counts().sort_index().plot(kind='line')
plt.title('Distribution of Patient Visits Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Visits')
plt.show()
```
